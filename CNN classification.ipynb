```python
def FinanceData(code,start = '2010-01-01',end = '2022-01-01'):
    df = fdr.DataReader(str(code), start, end)
    df = df.drop(['Change'],axis = 1)
    col = [i.lower() for i in df.columns.to_list()]
    df.columns = col
    return df
FinanceData('098120')
```
